# 고객별 리뷰 리스트로 변환

```python
import os
import pandas as pd
```

```python
directory_path = "/content/drive/MyDrive/DArt-B 24-2 학술제 유통물류/고객별 리뷰 내용"

data = []
processed_files = []
skipped_files = []

if os.path.exists(directory_path):
    for file_name in sorted(os.listdir(directory_path)):
        if file_name.endswith('.csv'): # 경로 내 csv 파일명에서 정보 가져옴
            original_file_name = file_name
            parts = file_name.replace('.csv', '').replace('#', '*').replace('_', ', ').split(', ')
            if len(parts) == 8:
                user_name = parts[0]
                helpful_count = parts[2]
                five_star = parts[3]
                four_star = parts[4]
                three_star = parts[5]
                two_star = parts[6]
                one_star = parts[7]

                file_path = os.path.join(directory_path, file_name)
                try:
                    df = pd.read_csv(file_path)

                    # 300행 이상인 경우 자르기
                    if len(df) > 300:
                        df = df.iloc[:300]

                    df['review_contents'] = df['review_contents'].fillna('').astype(str)
                    df['dates'] = df['dates'].fillna('').astype(str)

                    review_contents = df['review_contents'].tolist()
                    dates = df['dates'].tolist()

                    data.append([
                        original_file_name, user_name, helpful_count, five_star,
                        four_star, three_star, two_star, one_star, dates, review_contents
                    ])
                    processed_files.append(file_name)
                except Exception as e:
                    print(f"Error processing file {file_name}: {e}")
                    skipped_files.append(file_name)
            else:
                skipped_files.append(file_name)
```

```python
columns = [
    'original_file_name', 'user_name', 'helpful_count', '5_star', '4_star',
    '3_star', '2_star', '1_star', 'dates', 'review_contents'
]

final_df = pd.DataFrame(data, columns=columns)

# 처리된 파일 및 누락된 파일 수 출력
print(f"Processed files: {len(processed_files)}")
print(f"Skipped files: {len(skipped_files)}")

if skipped_files:
    print("Skipped files:")
    for skipped in skipped_files:
        print(skipped)

# 컬럼 순서 바꾸기
column_order = [
    'user_name', 'dates', 'review_contents', 'helpful_count',
    '5_star', '4_star', '3_star', '2_star', '1_star'
]

final_df = final_df[column_order]

output_file_path = os.path.join(directory_path, "(홈&주방)_200_수정.csv")
final_df.to_csv(output_file_path, index=False)

print(f"DataFrame saved to {output_file_path} with columns in specified order.")
```
